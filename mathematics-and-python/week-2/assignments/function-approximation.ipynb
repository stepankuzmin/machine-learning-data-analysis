# Линейная алгебра: сходство текстов и аппроксимация функций
## Аппроксимация функции

```python
import math
import numpy
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
f = lambda x: math.sin(x / 5.) * math.exp(x / 10.) + 5 * math.exp(-x / 2.)
```

```python
x = range(0, 16, 1)
y = map(f, x)
plt.plot(x, y)
```

```python
A = numpy.array([
        [1, 1],
        [1, 15]])

b = numpy.array([f(1), f(15)])

W1 = linalg.solve(A, b)
W1
```

```python
f1 = lambda x: 3.25221687 + 0.63522142 * x
y1 = map(f1, x)
plt.plot(x, y, x, y1)
```

```python
A2 = numpy.array([
    [1, 1, 1],
    [1, 8, 8**2],
    [1, 15, 15**2]])

b2 = numpy.array([f(1), f(8), f(15)])

W2 = linalg.solve(A2, b2)
W2
```

```python
f2 = lambda x: 3.32512949 -0.06531159 * x -0.00760104 * x **2
y2 = map(f2, x)
plt.plot(x, y, x, y2)
```

```python
A3 = numpy.array([
    [1, 1, 1, 1],
    [1, 4, 4**2, 4**3],
    [1, 10, 10**2, 10**3],
    [1, 15, 15**2, 15**3]])

b3 = numpy.array([f(1), f(4), f(10), f(15)])

W3 = linalg.solve(A3, b3)
print W3
```

```python
f3 = lambda x: 4.36264154 - 1.29552587 * x + 0.19333685 * x **2 - 0.00823565 * x**3
y3 = map(f3, x)
plt.plot(x, y, x, y3)
```
