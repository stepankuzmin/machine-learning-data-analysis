# Global optimization

```python
from math import sin, exp
from matplotlib.pyplot import plot
from scipy.optimize import differential_evolution
```

```python
f = lambda x: sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)
```

```python
x = range(1, 30)
y = map(f, x)
plot(x, y)
```

```python
bounds = [(1, 30)]
res = differential_evolution(f, bounds)
print res
print 'Differential evolution resulted with minimum at', f(res.x[0])

plot(x, y, res.x, f(res.x), 'ro')
```

```python
answer = f(res.x[0])
file = open('global-optimization.txt', 'w')
file.write("%.2f" % (answer))
file.close()
```
