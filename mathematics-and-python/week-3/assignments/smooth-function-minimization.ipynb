# Smooth function optimization

```python
from math import sin, exp
from matplotlib.pyplot import plot
from scipy.optimize import minimize
```

```python
f = lambda x: sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)
```

```python
x = range(1, 30)
y = map(f, x)
plot(x, y)
```

## Default

```python
res0 = minimize(f, [1])
print 'Default minimization starting from', x0[0], 'resulted with minimum at', f(res0.x[0])

res1 = minimize(f, [30])
print 'Default minimization starting from', x1[0], 'resulted with minimum at', f(res1.x[0])

plot(x, y, res0.x, f(res0.x), 'ro', res1.x, f(res1.x), 'ro')
```

## BFGS

```python
x2 = [2]
res2 = minimize(f, x2, method='BFGS')
print 'BFGS starting from', x2[0], 'resulted with minimum at', f(res2.x[0])

x3 = [30]
res3 = minimize(f, x3, method='BFGS')
print 'BFGS starting from', x3[0], 'resulted with minimum at', f(res3.x[0])

plot(x, y, res2.x, f(res2.x), 'ro', res3.x, f(res3.x), 'ro')
```

```python
answer1 = f(res2.x[0])
answer2 = f(res3.x[0])
file = open('smooth-function-minimization.txt', 'w')
file.write("%.2f %.2f" % (answer1, answer2))
file.close()
```
