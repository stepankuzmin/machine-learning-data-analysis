# Non smooth function minimization

```python
from math import sin, exp
from matplotlib.pyplot import plot
from scipy.optimize import minimize, differential_evolution
```

```python
f = lambda x: sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)
h = lambda x: int(f(x))
```

```python
x = range(1, 30)
y = map(h, x)
plot(x, y)
```

## BFGS

```python
x0 = [30]
res0 = minimize(h, x0, method='BFGS')
print 'BFGS starting from', x0[0], 'resulted with minimum at', h(res0.x[0])

plot(x, y, res0.x, h(res0.x), 'ro')
```

```python
bounds = [(1, 30)]
res1 = differential_evolution(h, bounds)
print 'Differential evolution resulted with minimum at', h(res1.x[0])

plot(x, y, res1.x, h(res1.x), 'ro')
```

```python
answer1 = h(res0.x[0])
answer2 = h(res1.x[0])

file = open('non-smooth-function-minimization.txt', 'w')
file.write("%.2f %.2f" % (answer1, answer2))
file.close()
```
